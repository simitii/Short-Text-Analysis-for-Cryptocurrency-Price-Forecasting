```python
from datetime import date, timedelta

data = []

start_date = date(2021, 1, 1)
end_date = date(2022, 5, 15)
delta = timedelta(days=1)
dates = []
while start_date <= end_date:
    current_day = start_date.strftime("%m%d%Y")
    start_date += delta
    dates.append(current_day)
```

```python
import requests
from tqdm.auto import tqdm
import time

cryptonews_TOKEN = "CRYPTONEWS TOKEN"

data = []

for current_day in tqdm(dates):
    # Get the first page of the day
    r = requests.get(f'https://cryptonews-api.com/api/v1/category?section=alltickers&items=50&page=1&date={current_day}-{current_day}&sortby=oldestfirst&token={cryptonews_TOKEN}')
    if r.status_code != 200:
        print("Error", current_day, 1, r.status_code, r)
        raise Exception()
    r_json = r.json()
    data += r_json['data']
    time.sleep(5)
    
    # Get next pages of the day
    for i in range(2, r_json['total_pages']+1):
        r = requests.get(f'https://cryptonews-api.com/api/v1/category?section=alltickers&items=50&page={i}&date={current_day}-{current_day}&sortby=oldestfirst&token={cryptonews_TOKEN}')
        if r.status_code != 200:
            print("Error", current_day, i, r.status_code, r)
            raise Exception()
        r_json = r.json()
        data += r_json['data']
        time.sleep(5)
```

```python
import pickle

with open('data.pickle', 'wb') as file:
    pickle.dump(data, file)
```

```python
symbols = []

for symbol_info in client.get_exchange_info()['symbols']:
    symbol = symbol_info['symbol']
    if symbol[-4:] == 'USDT':
        symbols.append(symbol)
```

```python
from datetime import datetime, timezone
import numpy as np

for i in tqdm(range(len(data))):
    coin = data[i]['tickers'][0]
    pair = coin+'USDT'
    if 'prices' in data[i]:
        del data[i]['prices']
    if 'volumes' in data[i]:
        del data[i]['volumes']
    try:
        date = datetime.strptime(data[i]['date'], '%a, %d %b %Y %H:%M:%S %z').astimezone(tz=timezone.utc)
        timestamp = date.timestamp() - date.second
        timestamp_before = timestamp - 6*3600 # 6 hours ago
        timestamp_after = timestamp + 6*3600 # 6 hour later

        kline_data = client.get_historical_klines(pair, Client.KLINE_INTERVAL_1MINUTE, int(timestamp_before*1000), int(timestamp_after*1000))
        assert len(kline_data) == 721 #12*60+1
        assert kline_data[0][0] == int(timestamp_before*1000)
        assert kline_data[360][0] == int(timestamp*1000)
        assert kline_data[-1][0] == int(timestamp_after*1000)

        kline_data = np.array(kline_data, dtype=float)
        data[i]['prices'] = {'before':kline_data[0:360,1], 'newstime':kline_data[360,1], 'after':kline_data[361:,1]}
        data[i]['volumes'] = {'before':kline_data[0:360,7], 'newstime':kline_data[360,7], 'after':kline_data[361:,7]}
    except Exception as e:
        print("Error (Pass)", i, coin, data[i]['date'])
        print(e)
```
